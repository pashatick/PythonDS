## РљР»Р°СЃС‚РµСЂРёР·Р°С†РёСЏ

РћРґРЅРёРј РёР· СЃР°РјС‹С… С‡Р°СЃС‚Рѕ РїСЂРёРјРµРЅСЏРµРјС‹С… РјРµС‚РѕРґРѕРІ РѕР±СѓС‡РµРЅРёСЏ Р±РµР· СѓС‡РёС‚РµР»СЏ СЏРІР»СЏРµС‚СЃСЏ _РєР»Р°СЃС‚РµСЂРёР·Р°С†РёСЏ_. Р—Р°РґР°С‡Р° РєР»Р°СЃС‚РµСЂРёР·Р°С†РёРё -  СЂР°Р·Р±РёС‚СЊ РѕР±СЉРµРєС‚С‹ РёР· РІС‹Р±РѕСЂРєРё РЅР° РѕРїСЂРµРґРµР»С‘РЅРЅРѕРµ С‡РёСЃР»Рѕ РіСЂСѓРїРї (РёР»Рё _РєР»Р°СЃС‚РµСЂРѕРІ_) С‚Р°Рє, С‡С‚РѕР±С‹ "РїРѕС…РѕР¶РёРµ" РѕР±СЉРµРєС‚С‹ РїРѕРїР°Р»Рё РІ РѕРґРёРЅ РєР»Р°СЃС‚РµСЂ. 

Р—РґРµСЃСЊ РјС‹ РїРѕР·РЅР°РєРѕРјРёРјСЃСЏ СЃ РѕРґРЅРёРј РёР· РЅР°РёР±РѕР»РµРµ РёР·РІРµСЃС‚РЅС‹С… Р°Р»РіРѕСЂРёС‚РјРѕРІ РєР»Р°СЃС‚РµСЂРёР·Р°С†РёРё - _k-means_ РёР»Рё _k СЃСЂРµРґРЅРёС…_.

```python
from sklearn.cluster import KMeans

model = KMeans(n_clusters=3, random_state=100)
```

Р”Р°РЅРЅР°СЏ РјРѕРґРµР»СЊ РїСЂРёРЅРёРјР°РµС‚ РЅР° РІС…РѕРґ РїР°СЂР°РјРµС‚СЂ `n_clusters`, С‚.Рµ. С‡РёСЃР»Рѕ РєР»Р°СЃС‚РµСЂРѕРІ, РЅР° РєРѕС‚РѕСЂС‹Рµ РјС‹ С…РѕС‚РµР»Рё Р±С‹ СЂР°Р·Р±РёС‚СЊ РЅР°С€Рё РґР°РЅРЅС‹Рµ.

РќР° РЅР°С‡Р°Р»СЊРЅРѕР№ РёС‚РµСЂР°С†РёРё Р°Р»РіРѕСЂРёС‚Рј РїСЂРѕРёР·РІРѕР»СЊРЅС‹Рј РѕР±СЂР°Р·РѕРј РІС‹Р±РёСЂР°РµС‚ С†РµРЅС‚СЂС‹ РєР»Р°СЃС‚РµСЂРѕРІ - _С†РµРЅС‚СЂРѕРёРґС‹_ (СЃС‚РѕР»СЊРєРѕ, СЃРєРѕР»СЊРєРѕ РјС‹ Р·Р°РґР°Р»Рё РІ РїР°СЂР°РјРµС‚СЂРµ `n_clusters`). Р—Р°С‚РµРј РїРѕРІС‚РѕСЂСЏРµС‚СЃСЏ СЃР»РµРґСѓСЋС‰РёР№ РїСЂРѕС†РµСЃСЃ:

1. РћС‚РЅРµСЃС‚Рё РєР°Р¶РґС‹Р№ РѕР±СЉРµРєС‚ РёР· РІС‹Р±РѕСЂРєРё Рє С‚РѕРјСѓ РєР»Р°СЃС‚РµСЂСѓ, С‡РµР№ С†РµРЅС‚СЂ Р±Р»РёР¶Рµ РІСЃРµРіРѕ Рє СЌС‚РѕРјСѓ РѕР±СЉРµРєС‚Сѓ.
2. РџРµСЂРµСЃС‚Р°РІРёС‚СЊ С†РµРЅС‚СЂРѕРёРґ РєР°Р¶РґРѕРіРѕ РєР»Р°СЃС‚РµСЂР° РІ С†РµРЅС‚СЂ СЌС‚РѕРіРѕ РєР»Р°СЃС‚РµСЂР° (СЃ СѓС‡С‘С‚РѕРј РЅРѕРІС‹С… РѕР±СЉРµРєС‚РѕРІ).

Р­С‚РѕС‚ РїСЂРѕС†РµСЃСЃ РїРѕРІС‚РѕСЂСЏРµС‚СЃСЏ РґРѕ С‚РµС… РїРѕСЂ, РїРѕРєР° РєРѕРѕСЂРґРёРЅР°С‚С‹ С†РµРЅС‚СЂРѕРІ РЅРµ РїРµСЂРµСЃС‚Р°РЅСѓС‚ РёР·РјРµРЅСЏС‚СЊСЃСЏ.

Р Р°Р·Р±РµСЂС‘РјСЃСЏ СЃ СЌС‚РёРј Р°Р»РіРѕСЂРёС‚РјРѕРј РЅР° РїСЂР°РєС‚РёРєРµ. Р”Р»СЏ СЌС‚РѕРіРѕ РІРѕСЃРїРѕР»СЊР·СѓРµРјСЃСЏ РіРѕС‚РѕРІРѕР№ С„СѓРЅРєС†РёРµР№ РґР»СЏ СЃРѕР·РґР°РЅРёСЏ РґР°С‚Р°СЃРµС‚Р°.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

plt.style.use('fivethirtyeight')

%config InlineBackend.figure_format = 'svg'
%matplotlib inline
```

```python
def make_data(n, seed):
    np.random.seed(seed)
    
    shift_matrix = np.array([[3, 3], 
                             [6, 9], 
                             [9, 3]])
    
    data = np.random.randn(3, 2, n) + shift_matrix.reshape((3, 2, 1))
    data = np.swapaxes(data, 1, 2)
    data = data.reshape((-1, 2))
    data *= np.array([[20, 0.5]])
    
    df = pd.DataFrame({'x': data[:, 0], 'y': data[:, 1]},
                      columns=['x', 'y'])
    df = df.sample(frac=1.0)

    return df
```

Р­С‚Р° С„СѓРЅРєС†РёСЏ РїСЂРёРЅРёРјР°РµС‚ РЅР° РІС…РѕРґ РїР°СЂР°РјРµС‚СЂС‹ `n` (С‡РёСЃР»Рѕ РѕР±СЉРµРєС‚РѕРІ РІ РєР°Р¶РґРѕР№ РіСЂСѓРїРїРµ РѕР±СЉРµРєС‚РѕРІ) Рё `seed` (Р·РµСЂРЅРѕ, РЅРµРѕР±С…РѕРґРёРјРѕРµ РґР»СЏ РІРѕСЃРїСЂРѕРёР·РІРѕРґРёРјРѕСЃС‚Рё СЃР»СѓС‡Р°Р№РЅС‹С… Р·РЅР°С‡РµРЅРёР№). Р’ СЂРµР·СѓР»СЊС‚Р°С‚Рµ РїРѕР»СѓС‡Р°РµС‚СЃСЏ `DataFrame`, СЃРѕРґРµСЂР¶Р°С‰РёР№ Р·Р°РґР°РЅРЅРѕРµ С‡РёСЃР»Рѕ РѕР±СЉРµРєС‚РѕРІ РІ РєР°Р¶РґРѕР№ РіСЂСѓРїРїРµ (РІСЃРµРіРѕ 3 РіСЂСѓРїРїС‹) СЃ РґРІСѓРјСЏ РїСЂРёР·РЅР°РєР°РјРё: `'x'` Рё `'y'`.

РџРѕР»СѓС‡РёРј СЃ РїРѕРјРѕС‰СЊСЋ СЌС‚РѕР№ С„СѓРЅРєС†РёРё РґРІР° РЅР°Р±РѕСЂР° РґР°РЅРЅС‹С…. 

```python
train = make_data(10, 42)
test = make_data(10, 27)
```

Р§С‚РѕР±С‹ РїСЂРёРґР°С‚СЊ РґР°РЅРЅС‹Рј Р±РѕР»СЊС€РёР№ РїСЂРµРґРјРµС‚РЅС‹Р№ СЃРјС‹СЃР», РїСЂРµРґСЃС‚Р°РІРёРј, С‡С‚Рѕ РїРѕР»СѓС‡РµРЅРЅС‹Рµ РЅР°РјРё РґР°РЅРЅС‹Рµ - СЌС‚Рѕ РґР°РЅРЅС‹Рµ Рѕ СЂР°Р·РЅС‹С… РІРёРґР°С… СЂС‹Р±, РѕР±РёС‚Р°СЋС‰РёС… РІ РІРѕРґРѕС‘РјРµ. РњС‹ Р±С‹ С…РѕС‚РµР»Рё СЂР°Р·Р±РёС‚СЊ РїСЂРµРґСЃС‚Р°РІРёС‚РµР»РµР№ СЂС‹Р± РёР· СЌС‚РѕРіРѕ РІРѕРґРѕС‘РјР° РЅР° 3 РєР»Р°СЃС‚РµСЂР°.

РџРѕСЃРјРѕС‚СЂРёРј, РєР°Рє СЂР°СЃРїРѕР»РѕР¶РµРЅС‹ РЅР°С€Рё РѕР±СЉРµРєС‚С‹ РІ РїСЂРѕСЃС‚СЂР°РЅСЃС‚РІРµ. РџРѕСЃРєРѕР»СЊРєСѓ Сѓ РєР°Р¶РґРѕРіРѕ РѕР±СЉРµРєС‚Р° РІСЃРµРіРѕ РґРІР° РїСЂРёР·РЅР°РєР°, РјС‹ РјРѕР¶РµРј РїСЂРµРґСЃС‚Р°РІРёС‚СЊ РєР°Р¶РґС‹Р№ РѕР±СЉРµРєС‚ РєР°Рє С‚РѕС‡РєСѓ РЅР° РїР»РѕСЃРєРѕСЃС‚Рё:

```python
plt.scatter(train['x'], train['y'])

plt.xlabel('Р”Р»РёРЅР° С‚РµР»Р°')
plt.ylabel('Р”Р»РёРЅР° С‡РµС€СѓРё')

plt.title('Train data')

plt.show()
```

РџСЂРµР¶РґРµ С‡РµРј РїСЂРёСЃС‚СѓРїРёС‚СЊ Рє РєР»Р°СЃС‚РµСЂРёР·Р°С†РёРё, РѕС‚РјР°СЃС€С‚Р°Р±РёСЂСѓРµРј РЅР°С€Рё РґР°РЅРЅС‹Рµ СЃ РїРѕРјРѕС‰СЊСЋ СѓР¶Рµ Р·РЅР°РєРѕРјРѕРіРѕ РЅР°Рј `MinMaxScaler`:

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()

train_scaled = pd.DataFrame(scaler.fit_transform(train), columns=['x', 'y'])

plt.scatter(train_scaled['x'], train_scaled['y'])

plt.xlabel('Р”Р»РёРЅР° С‚РµР»Р°')
plt.ylabel('Р”Р»РёРЅР° С‡РµС€СѓРё')

plt.title('Train data (scaled)')

plt.show()
```

РС‚Р°Рє, РїРµСЂРµР№РґС‘Рј Рє РєР»Р°СЃС‚РµСЂРёР·Р°С†РёРё. РЎ РїРѕРјРѕС‰СЊСЋ РјРµС‚РѕРґР° `.fit_predict` РјРѕР¶РЅРѕ РѕР±СѓС‡РёС‚СЊ РјРѕРґРµР»СЊ, Р° Р·Р°С‚РµРј СЃСЂР°Р·Сѓ Р¶Рµ РїРѕР»СѓС‡РёС‚СЊ РјРµС‚РєРё РєР»Р°СЃС‚РµСЂРѕРІ РЅР° РІС…РѕРґРЅС‹С… РґР°РЅРЅС‹С…:

```python
train_labels = model.fit_predict(train_scaled)

train_labels
```

```python
centers = scaler.inverse_transform(model.cluster_centers_)
centers
```

РЎ РїРѕРјРѕС‰СЊСЋ Р°С‚СЂРёР±СѓС‚Р° `.cluster_centers_` РјРѕР¶РЅРѕ РїРѕР»СѓС‡РёС‚СЊ РїРѕР»СѓС‡РёРІС€РёРµСЃСЏ РІ СЂРµР·СѓР»СЊС‚Р°С‚Рµ СЂР°Р±РѕС‚С‹ Р°Р»РіРѕСЂРёС‚РјР° С†РµРЅС‚СЂС‹ РєР»Р°СЃС‚РµСЂРѕРІ. Р§С‚РѕР±С‹ РёР·РѕР±СЂР°Р·РёС‚СЊ СЌС‚Рё С†РµРЅС‚СЂС‹ РЅР° РїР»РѕСЃРєРѕСЃС‚Рё, РІРѕСЃРїРѕР»СЊР·СѓРµРјСЃСЏ РјРµС‚РѕРґРѕРј `scaler.inverse_transform`, РєРѕС‚РѕСЂС‹Р№ РїСЂРµРѕР±СЂР°Р·СѓРµС‚ РѕС‚РјР°СЃС€С‚Р°Р±РёСЂРѕРІР°РЅРЅС‹Рµ Р·РЅР°С‡РµРЅРёСЏ РѕР±СЂР°С‚РЅРѕ РІ СЃС‹СЂС‹Рµ. Р¦РµРЅС‚СЂС‹ РЅР° РіСЂР°С„РёРєРµ РїРѕРјРµС‚РёРј РєСЂР°СЃРЅС‹РјРё СЂРѕРјР±Р°РјРё. РљСЂРѕРјРµ С‚РѕРіРѕ, РїРѕРєСЂР°СЃРёРј РѕР±СЉРµРєС‚С‹ РёР· СЂР°Р·РЅС‹С… РєР»Р°СЃС‚РµСЂРѕРІ РІ СЂР°Р·РЅС‹Рµ С†РІРµС‚Р° РґР»СЏ РЅР°РіР»СЏРґРЅРѕСЃС‚Рё.

```python
centers = scaler.inverse_transform(model.cluster_centers_)

plt.scatter(train['x'], train['y'], c=train_labels)
plt.scatter(centers[:, 0], centers[:, 1], marker='D', color='red')

plt.xlabel('Р”Р»РёРЅР° С‚РµР»Р°')
plt.ylabel('Р”Р»РёРЅР° С‡РµС€СѓРё')

plt.title('Train data')
```

РћСЃРЅРѕРІРЅР°СЏ РїСЂРѕР±Р»РµРјР° Р°Р»РіРѕСЂРёС‚РјР° k-means РІ С‚РѕРј, С‡С‚Рѕ РЅР°Рј РЅСѓР¶РЅРѕ Р·Р°СЂР°РЅРµРµ Р·РЅР°С‚СЊ РєРѕР»РёС‡РµСЃС‚РІРѕ РєР»Р°СЃС‚РµСЂРѕРІ. РћРґРЅР°РєРѕ, РІ СЂРµР°Р»СЊРЅРѕСЃС‚Рё СЌС‚Рѕ РЅРµ РІСЃРµРіРґР° Р±С‹РІР°РµС‚ РІРѕР·РјРѕР¶РЅРѕ. РљРѕР»РёС‡РµСЃС‚РІРѕ РєР»Р°СЃС‚РµСЂРѕРІ РјРѕР¶РЅРѕ РѕС†РµРЅРёС‚СЊ, РёСЃРїРѕР»СЊР·СѓСЏ РІРµР»РёС‡РёРЅСѓ _inertia_. Р­С‚Р° РІРµР»РёС‡РёРЅР° СЂР°РІРЅР° СЃСѓРјРјРµ РєРІР°РґСЂР°С‚РѕРІ СЂР°СЃСЃС‚РѕСЏРЅРёР№ РѕС‚ РѕР±СЉРµРєС‚РѕРІ РґРѕ С†РµРЅС‚СЂРѕРІ РёС… РєР»Р°СЃС‚РµСЂРѕРІ. Р§РµРј Р»СѓС‡С€Рµ РєР»Р°СЃС‚РµСЂРёР·Р°С†РёСЏ, С‚РµРј РјРµРЅСЊС€Рµ СЌС‚Р° РІРµР»РёС‡РёРЅР°. РљСЂРѕРјРµ С‚РѕРіРѕ, РѕРЅР° СѓРјРµРЅСЊС€Р°РµС‚СЃСЏ СЃ СѓРІРµР»РёС‡РµРЅРёРµРј С‡РёСЃР»Р° РєР»Р°СЃС‚РµСЂРѕРІ (РѕРЅР° СЂР°РІРЅР° 0, РµСЃР»Рё РєР»Р°СЃС‚РµСЂРѕРІ СЃС‚РѕР»СЊРєРѕ Р¶Рµ, СЃРєРѕР»СЊРєРѕ РѕР±СЉРµРєС‚РѕРІ), РїРѕСЌС‚РѕРјСѓ РЅСѓР¶РЅРѕ СЃРѕР±Р»СЋРґР°С‚СЊ Р±Р°Р»Р°РЅСЃ РјРµР¶РґСѓ РєРѕР»РёС‡РµСЃС‚РІРѕРј РєР»Р°СЃС‚РµСЂРѕРІ, РєРѕС‚РѕСЂРѕРµ РЅР°Рј РїРѕРґС…РѕРґРёС‚, Рё РІРµР»РёС‡РёРЅРѕР№ inertia, С‡С‚РѕР±С‹ С‡РёСЃР»Рѕ РєР»Р°СЃС‚РµСЂРѕРІ Р±С‹Р»Рѕ СЂР°Р·СѓРјРЅС‹Рј.

## РџРѕРЅРёР¶РµРЅРёРµ СЂР°Р·РјРµСЂРЅРѕСЃС‚Рё РґР°РЅРЅС‹С…

РџСЂРµРґРїРѕР»РѕР¶РёРј, РїРµСЂРµРґ РЅР°РјРё СЃС‚РѕРёС‚ Р·Р°РґР°С‡Р° РѕР±СѓС‡РµРЅРёСЏ СЃ СѓС‡РёС‚РµР»РµРј, Рё РІ РЅР°С€РёС… РґР°РЅРЅС‹С… РѕС‡РµРЅСЊ РјРЅРѕРіРѕ РїСЂРёР·РЅР°РєРѕРІ (СЃРѕС‚РЅРё, С‚С‹СЃСЏС‡Рё РёР»Рё РґР°Р¶Рµ Р±РѕР»СЊС€Рµ). РћРґРёРЅ РёР· С‚Р°РєРёС… РјРµС‚РѕРґРѕРІ - _РњРµС‚РѕРґ РіР»Р°РІРЅС‹С… РєРѕРјРїРѕРЅРµРЅС‚_ (_Principal component analysis_ РёР»Рё _PCA_). Р”Р»СЏ РЅР°С…РѕР¶РґРµРЅРёСЏ РіР»Р°РІРЅС‹С… РєРѕРјРїРѕРЅРµРЅС‚ РЅСѓР¶РЅРѕ РІС‹С‡РёСЃР»РёС‚СЊ РєРѕРІР°СЂРёР°С†РёРѕРЅРЅСѓСЋ РјР°С‚СЂРёС†Сѓ, Р·Р°С‚РµРј РґР»СЏ СЌС‚РѕР№ РјР°С‚СЂРёС†С‹ РЅР°Р№С‚Рё СЃРѕР±СЃС‚РІРµРЅРЅС‹Рµ РІРµРєС‚РѕСЂС‹ Рё СЃРѕР±СЃС‚РІРµРЅРЅС‹Рµ Р·РЅР°С‡РµРЅРёСЏ. РџРµСЂРµРґ РїСЂРёРјРµРЅРµРЅРёРµРј РґР°РЅРЅРѕРіРѕ РјРµС‚РѕРґР° РІР°Р¶РЅРѕ РЅРµ Р·Р°Р±С‹С‚СЊ РїСЂРёРІРµСЃС‚Рё РІСЃРµ РґР°РЅРЅС‹Рµ Рє СЃС…РѕР¶РµРјСѓ РјР°СЃС€С‚Р°Р±Сѓ. Р­С‚Рѕ РјРѕР¶РЅРѕ СЃРґРµР»Р°С‚СЊ, РЅР°РїСЂРёРјРµСЂ, СЃ РїРѕРјРѕС‰СЊСЋ СЃС‚Р°РЅРґР°СЂС‚РёР·Р°С†РёРё РёР»Рё РЅРѕСЂРјР°Р»РёР·Р°С†РёРё.

### РџСЂРёРјРµРЅРµРЅРёРµ PCA РІ РјРѕРґРµР»СЏС… РјР°С€РёРЅРЅРѕРіРѕ РѕР±СѓС‡РµРЅРёСЏ

РџРѕРїСЂРѕР±СѓРµРј РїСЂРёРјРµРЅРёС‚СЊ РјРµС‚РѕРґ РіР»Р°РІРЅС‹С… РєРѕРјРїРѕРЅРµРЅС‚ РІ РјРѕРґРµР»Рё РєР»Р°СЃСЃРёС„РёРєР°С†РёРё. Р—Р°РіСЂСѓР·РёРј РґР°РЅРЅС‹Рµ РїРѕ РїР°СЃСЃР°Р¶РёСЂР°Рј РўРёС‚Р°РЅРёРєР°, СЃРѕС…СЂР°РЅС‘РЅРЅС‹Рµ СЂР°РЅРµРµ:

```python
X_train = pd.read_pickle('X_train.pkl')
y_train = pd.read_pickle('y_train.pkl')

X_valid = pd.read_pickle('X_valid.pkl')
y_valid = pd.read_pickle('y_valid.pkl')
```

Р‘СѓРґРµРј РјР°СЃС€С‚Р°Р±РёСЂРѕРІР°С‚СЊ РґР°РЅРЅС‹Рµ СЃ РїРѕРјРѕС‰СЊСЋ СЃС‚Р°РЅРґР°СЂС‚РёР·Р°С†РёРё. РџРѕРґР°РґРёРј СЃС‚Р°РЅРґР°СЂС‚РёР·Р°С‚РѕСЂСѓ РїР°СЂР°РјРµС‚СЂ `with_mean=False` С‡С‚РѕР±С‹ РЅРµ РїСЂРѕРІРѕРґРёС‚СЊ С†РµРЅС‚СЂРёСЂРѕРІР°РЅРёРµ.

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler(with_mean=False)

X_train_scaled = pd.DataFrame(scaler.fit_transform(X_train), columns=X_train.columns)
X_valid_scaled = pd.DataFrame(scaler.transform(X_valid), columns=X_valid.columns)
```

Р—Р°РїСѓСЃС‚РёРј С‚РµРїРµСЂСЊ РјРµС‚РѕРґ РіР»Р°РІРЅС‹С… РєРѕРјРїРѕРЅРµРЅС‚. Р—РґРµСЃСЊ РјС‹ РЅРµ Р±СѓРґРµРј СѓРєР°Р·С‹РІР°С‚СЊ С‡РёСЃР»Рѕ РіР»Р°РІРЅС‹С… РєРѕРјРїРѕРЅРµРЅС‚, РїРѕ СѓРјРѕР»С‡Р°РЅРёСЋ РјРѕРґРµР»СЊ `PCA` РїРѕСЃС‡РёС‚Р°РµС‚ РІСЃРµ. РџРѕСЃС‡РёС‚Р°РµРј РіР»Р°РІРЅС‹Рµ РєРѕРјРїРѕРЅРµРЅС‚С‹ Рё РїРѕСЃРјРѕС‚СЂРёРј РЅР° РґРѕР»СЋ РѕР±СЉСЏСЃРЅС‘РЅРЅРѕР№ РґРёСЃРїРµСЂСЃРёРё РєР°Р¶РґРѕР№ РєРѕРјРїРѕРЅРµРЅС‚С‹.

```python
pca = PCA(random_state=100)

pca.fit(X_train_scaled)

print(pca.explained_variance_ratio_)
```

РџРѕСЃС‡РёС‚Р°РµРј СЃСѓРјРјСѓ РїРµСЂРІС‹С… 7 Р·РЅР°С‡РµРЅРёР№:

```python
pca.explained_variance_ratio_[:7].sum()
```

Р•С‰С‘ СЂР°Р· СЃРѕР·РґР°РґРёРј РјРѕРґРµР»СЊ `PCA`, РЅР° СЌС‚РѕС‚ СЂР°Р· СѓРєР°Р·Р°РІ С‡РёСЃР»Рѕ РєРѕРјРїРѕРЅРµРЅС‚, Рё РїРѕР»СѓС‡РёРј С‚СЂРµРЅРёСЂРѕРІРѕС‡РЅС‹Рµ Рё РІР°Р»РёРґР°С†РёРѕРЅРЅС‹Рµ РґР°РЅРЅС‹Рµ РїРѕРЅРёР¶РµРЅРЅРѕР№ СЂР°Р·РјРµСЂРЅРѕСЃС‚Рё:

```python
pca = PCA(n_components=7, random_state=100)

X_train_mc = pca.fit_transform(X_train_scaled)
X_valid_mc = pca.transform(X_valid_scaled)
```

РўРµРїРµСЂСЊ РѕР±СѓС‡РёРј РЅР° СЌС‚РёС… РґР°РЅРЅС‹С… РјРѕРґРµР»СЊ Р»РѕРіРёСЃС‚РёС‡РµСЃРєРѕР№ СЂРµРіСЂРµСЃСЃРёРё.

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

lr = LogisticRegression()

lr.fit(X_train_mc, y_train)

y_pred = lr.predict(X_valid_mc)

print(accuracy_score(y_valid, y_pred))
```

## РђР»РіРѕСЂРёС‚Рј t-SNE

РђР»РіРѕСЂРёС‚Рј _t-SNE_ (_t-distributed Stochastic Neighbor Embedding_ РёР»Рё _РЎС‚РѕС…Р°СЃС‚РёС‡РµСЃРєРѕРµ РІР»РѕР¶РµРЅРёРµ СЃРѕСЃРµРґРµР№ СЃ t-СЂР°СЃРїСЂРµРґРµР»РµРЅРёРµРј_) РїРѕР·РІРѕР»СЏРµС‚ РїРѕРЅРёР¶Р°С‚СЊ СЂР°Р·РјРµСЂРЅРѕСЃС‚СЊ РґР°РЅРЅС‹С… РґРѕ РґРІСѓС… РёР»Рё С‚СЂС‘С… РёР·РјРµСЂРµРЅРёР№, С‡С‚Рѕ РїРѕР·РІРѕР»СЏРµС‚ РІРёР·СѓР°Р»РёР·РёСЂРѕРІР°С‚СЊ РґР°РЅРЅС‹Рµ РЅР° РґРІСѓРјРµСЂРЅС‹С… Рё С‚СЂС‘С…РјРµСЂРЅС‹С… РіСЂР°С„РёРєР°С…. РР·СѓС‡Р°СЏ РіСЂР°С„РёРєРё, РјРѕР¶РЅРѕ, РЅР°РїСЂРёРјРµСЂ, РїРѕРЅСЏС‚СЊ, РЅР° СЃРєРѕР»СЊРєРѕ РєР»Р°СЃС‚РµСЂРѕРІ Р°РґРµРєРІР°С‚РЅРѕ СЂР°Р·Р±РёРІР°С‚СЊ РґР°РЅРЅС‹Рµ, Р° С‚Р°РєР¶Рµ РѕС†РµРЅРёС‚СЊ СѓР¶Рµ РІС‹РїРѕР»РЅРµРЅРЅРѕРµ СЂР°Р·Р±РёРµРЅРёРµ РЅР° РєР»Р°СЃС‚РµСЂС‹.

Р Р°Р·Р±РµСЂС‘РјСЃСЏ РІ СЂР°Р±РѕС‚Рµ СЌС‚РѕРіРѕ Р°Р»РіРѕСЂРёС‚РјР° СЃ РїРѕРјРѕС‰СЊСЋ РґР°С‚Р°СЃРµС‚Р° Boston РёР· Р±РёР±Р»РёРѕС‚РµРєРё `sklearn`.

```python
from sklearn.manifold import TSNE
from sklearn.datasets import load_boston

boston = load_boston()

X = pd.DataFrame(boston.data, columns=boston.feature_names)
y = boston.target

X.info()
```

Р’РёРґРЅРѕ, С‡С‚Рѕ РґР°РЅРЅС‹Рµ РїСЂРµРґСЃС‚Р°РІР»СЏСЋС‚ РёР· СЃРµР±СЏ 13 СЃС‚РѕР»Р±С†РѕРІ СЃ РґСЂРѕР±РЅС‹РјРё Р·РЅР°С‡РµРЅРёСЏРјРё, РїСЂРѕРїСѓСЃРєРѕРІ РЅРµС‚.

Р Р°Р·РѕР±СЊС‘Рј РґР°РЅРЅС‹Рµ РЅР° С‚СЂРµРЅРёСЂРѕРІРѕС‡РЅС‹Р№ Рё С‚РµСЃС‚РѕРІС‹Р№ РґР°С‚Р°СЃРµС‚С‹.

```python
scaler = StandardScaler()

X_train_scaled = pd.DataFrame(scaler.fit_transform(X_train), columns=X_train.columns)
X_test_scaled = pd.DataFrame(scaler.transform(X_test), columns=X_test.columns)
```

РС‚Р°Рє, РїРµСЂРµР№РґС‘Рј Рє РёСЃРїРѕР»СЊР·РѕРІР°РЅРёСЋ t-SNE. Р—Р°РґР°РґРёРј РїР°СЂР°РјРµС‚СЂ `n_components=2`, С‡С‚РѕР±С‹ РїРѕР»СѓС‡РёС‚СЊ РґР°РЅРЅС‹Рµ СЃ РґРІСѓРјСЏ РїСЂРёР·РЅР°РєР°РјРё. РџР°СЂР°РјРµС‚СЂ `learning_rate` РІР»РёСЏРµС‚ РЅР° С‚Рѕ, РєР°Рє РїР»РѕС‚РЅРѕ Р±СѓРґСѓС‚ СЂР°СЃРїРѕР»Р°РіР°С‚СЊСЃСЏ С‚РѕС‡РєРё. Р РµРєРѕРјРµРЅРґСѓРµС‚СЃСЏ Р·Р°РґР°РІР°С‚СЊ РµРіРѕ РІ РґРёР°РїР°Р·РѕРЅРµ РѕС‚ 10 РґРѕ 1000.

```python
tsne = TSNE(n_components=2, learning_rate=150, random_state=100)

X_train_tsne = tsne.fit_transform(X_train_scaled)

print('Р”Рѕ:\t{}'.format(X_train_scaled.shape))
print('РџРѕСЃР»Рµ:\t{}'.format(X_train_tsne.shape))
```

РњС‹ РІРёРґРёРј, С‡С‚Рѕ С‡РёСЃР»Рѕ РїСЂРёР·РЅР°РєРѕРІ СѓРјРµРЅСЊС€РёР»РѕСЃСЊ СЃ 13 РґРѕ 2. РўРµРїРµСЂСЊ РјРѕР¶РЅРѕ РІРёР·СѓР°Р»РёР·РёСЂРѕРІР°С‚СЊ РЅР°С€Рё РґР°РЅРЅС‹Рµ РЅР° РїР»РѕСЃРєРѕСЃС‚Рё.

```python
plt.scatter(X_train_tsne[:, 0], X_train_tsne[:, 1])

plt.show()
```

РџРѕ РіСЂР°С„РёРєСѓ РІРёРґРЅРѕ, С‡С‚Рѕ РґР°РЅРЅС‹Рµ РјРѕР¶РЅРѕ СЂР°Р·Р±РёС‚СЊ РєР°Рє РјРёРЅРёРјСѓРј РЅР° 2 РєР»Р°СЃС‚РµСЂР°. РџРѕРїСЂРѕР±СѓРµРј СЃРґРµР»Р°С‚СЊ СЌС‚Рѕ СЃ РїРѕРјРѕС‰СЊСЋ СѓР¶Рµ РёР·РІРµСЃС‚РЅРѕРіРѕ РЅР°Рј РјРµС‚РѕРґР° K-means Рё РµС‰С‘ СЂР°Р· РїРѕСЃС‚СЂРѕРёРј РіСЂР°С„РёРє, РЅРѕ СѓР¶Рµ СЃ РїРѕР»СѓС‡РµРЅРЅС‹РјРё РјРµС‚РєР°РјРё РєР»Р°СЃС‚РµСЂРѕРІ.

```python
kmeans = KMeans(n_clusters=2)

labels_train = kmeans.fit_predict(X_train_scaled)

plt.scatter(X_train_tsne[:, 0], X_train_tsne[:, 1], c=labels_train)

plt.show()
```

РљР°Рє РјС‹ РІРёРґРёРј, РїСЂРё РєР»Р°СЃС‚РµСЂРёР·Р°С†РёРё РјС‹ РїРѕР»СѓС‡РёР»Рё РїСЂР°РєС‚РёС‡РµСЃРєРё С‚Р°РєРѕРµ Р¶Рµ СЂР°Р·Р±РёРµРЅРёРµ, РєР°Рє С‚Рѕ, РєРѕС‚РѕСЂРѕРµ РјРѕР¶РЅРѕ Р±С‹Р»Рѕ РЅР°Р±Р»СЋРґР°С‚СЊ РІ СЂРµР·СѓР»СЊС‚Р°С‚Рµ t-SNE.

Р”Р»СЏ С‚РµСЃС‚РѕРІРѕР№ РІС‹Р±РѕСЂРєРё РїРѕР»СѓС‡РёС‚СЊ Р°РЅР°Р»РѕРіРёС‡РЅС‹Р№ РіСЂР°С„РёРє РЅР°Рј РЅРµ СѓРґР°СЃС‚СЃСЏ, РїРѕСЃРєРѕР»СЊРєСѓ Сѓ Р°Р»РіРѕСЂРёС‚РјР° t-SNE РЅРµС‚ РјРµС‚РѕРґР° `.transform`. РћРґРЅР°РєРѕ, СЃ РїРѕРјРѕС‰СЊСЋ Р°Р»РіРѕСЂРёС‚РјР° K-means РјС‹ РјРѕР¶РµРј РїРѕР»СѓС‡РёС‚СЊ РјРµС‚РєРё РєР»Р°СЃС‚РµСЂРѕРІ РґР»СЏ С‚РµСЃС‚РѕРІРѕР№ РІС‹Р±РѕСЂРєРё СЃ РїРѕРјРѕС‰СЊСЋ РјРµС‚РѕРґР° `.predict` Рё РёСЃРїРѕР»СЊР·РѕРІР°С‚СЊ РёС…:

```python
labels_test = kmeans.predict(X_test_scaled)
```

РќР°РїСЂРёРјРµСЂ, РїРѕСЃРјРѕС‚СЂРёРј РЅР° СЃСЂРµРґРЅСЋСЋ С†РµРЅСѓ РЅРµРґРІРёР¶РёРјРѕСЃС‚Рё РІРѕ РІСЃРµР№ РІС‹Р±РѕСЂРєРµ Рё РІ РѕС‚РґРµР»СЊРЅС‹С… РєР»Р°СЃС‚РµСЂР°С….

РџРѕСЃРјРѕС‚СЂРёРј С‚РµРїРµСЂСЊ РЅР° СЂР°СЃРїСЂРµРґРµР»РµРЅРёРµ РѕС‚РґРµР»СЊРЅС‹С… РїСЂРёР·РЅР°РєРѕРІ РІРЅСѓС‚СЂРё РєР»Р°СЃС‚РµСЂРѕРІ. РќР°РїСЂРёРјРµСЂ, `'RM'` - РєРѕР»РёС‡РµСЃС‚РІРѕ РєРѕРјРЅР°С‚.

РћРЅ РѕС‚СЂР°Р¶Р°РµС‚ РєСЂРёРјРёРЅР°Р»СЊРЅРѕСЃС‚СЊ РјРµСЃС‚Р°, РІ РєРѕС‚РѕСЂРѕРј СЂР°СЃРїРѕР»РѕР¶РµРЅР° РЅРµРґРІРёР¶РёРјРѕСЃС‚СЊ (РєРѕР»РёС‡РµСЃС‚РІРѕ РїСЂРµСЃС‚СѓРїР»РµРЅРёР№ РЅР° РґСѓС€Сѓ РЅР°СЃРµР»РµРЅРёСЏ).

Р’ РєР»Р°СЃС‚РµСЂРµ 0 РјР°РєСЃРёРјР°Р»СЊРЅР°СЏ РїР»РѕС‚РЅРѕСЃС‚СЊ СЌС‚РѕРіРѕ РїСЂРёР·РЅР°РєР° СЂР°СЃРїРѕР»РѕР¶РµРЅР° РІРѕР·Р»Рµ 0. Р”Р»СЏ РєР»Р°СЃС‚РµСЂР° 1 Р·РЅР°С‡РµРЅРёСЏ СЌС‚РѕРіРѕ РїСЂРёР·РЅР°РєР° Р±РѕР»РµРµ СЂР°Р·РЅРѕРѕР±СЂР°Р·РЅС‹.

Р•С‰С‘ РѕРґРёРЅ РёРЅС‚РµСЂРµСЃРЅС‹Р№ РїСЂРёР·РЅР°Рє - РєРѕРЅС†РµРЅС‚СЂР°С†РёСЏ РѕРєСЃРёРґР° Р°Р·РѕС‚Р° РІ РІРѕР·РґСѓС…Рµ (`'NOX'`). РџРѕСЃС‚СЂРѕРёРј РіРёСЃС‚РѕРіСЂР°РјРјС‹ СЂР°СЃРїСЂРµРґРµР»РµРЅРёСЏ СЌС‚РѕРіРѕ РїСЂРёР·РЅР°РєР°.

Р’РЅРѕРІСЊ РєР»Р°СЃС‚РµСЂ 1 РІ СЌС‚РѕРј РїР»Р°РЅРµ РѕС‚Р»РёС‡Р°РµС‚СЃСЏ РІ С…СѓРґС€СѓСЋ СЃС‚РѕСЂРѕРЅСѓ. 

РС‚Р°Рє, РєР°Рє РЅР°Рј С‚РµРїРµСЂСЊ РёСЃРїРѕР»СЊР·РѕРІР°С‚СЊ РїРѕР»СѓС‡РµРЅРЅСѓСЋ РёРЅС„РѕСЂРјР°С†РёСЋ? РњС‹ РјРѕР¶РµРј, РЅР°РїСЂРёРјРµСЂ, РїРѕРїСЂРѕР±РѕРІР°С‚СЊ РїРѕСЃС‚СЂРѕРёС‚СЊ РѕС‚РґРµР»СЊРЅСѓСЋ РјРѕРґРµР»СЊ РґР»СЏ РєР°Р¶РґРѕРіРѕ РєР»Р°СЃС‚РµСЂР°. РЎРЅР°С‡Р°Р»Р° РїРѕСЃС‚СЂРѕРёРј РјРѕРґРµР»СЊ РЅР° РІСЃРµС… РґР°РЅРЅС‹С…. Р”Р»СЏ РѕС†РµРЅРєРё РєР°С‡РµСЃС‚РІР° РјРѕРґРµР»Рё Р±СѓРґРµРј РёСЃРїРѕР»СЊР·РѕРІР°С‚СЊ `r2_score`.

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score

lr = LinearRegression()

lr.fit(X_train_scaled, y_train)

y_test_pred = lr.predict(X_test_scaled)

r2_score(y_test, y_test_pred)
```

РС‚Р°Рє, РјС‹ РїРѕР»СѓС‡РёР»Рё Р±Р°Р·РѕРІРѕРµ Р·РЅР°С‡РµРЅРёРµ. РџРѕСЃС‚СЂРѕРёРј С‚РµРїРµСЂСЊ СЂР°Р·РЅС‹Рµ РјРѕРґРµР»Рё РґР»СЏ РѕР±СЉРµРєС‚РѕРІ СЂР°Р·РЅС‹С… РєР»Р°СЃС‚РµСЂРѕРІ Рё РїРѕСЃРјРѕС‚СЂРёРј, РєР°РєРѕРµ РєР°С‡РµСЃС‚РІРѕ РјС‹ СЃРјРѕР¶РµРј РїРѕР»СѓС‡РёС‚СЊ РЅР° РЅРёС….

РњРѕРґРµР»СЊ РґР»СЏ РєР»Р°СЃС‚РµСЂР° 0:

```python
lr.fit(X_train_scaled.loc[labels_train == 0], y_train[labels_train == 0])

y_test_pred_0 = lr.predict(X_test_scaled.loc[labels_test == 0])

r2_score(y_test[labels_test == 0], y_test_pred_0)
```

Р”Р»СЏ РєР»Р°СЃС‚РµСЂР° 0 РјС‹ РїРѕР»СѓС‡РёР»Рё Р±РѕР»СЊС€РµРµ Р·РЅР°С‡РµРЅРёРµ `r2_score`, РѕРґРЅР°РєРѕ, СЃСЂР°РІРЅРёРІР°С‚СЊ СЌС‚Рё РјРѕРґРµР»Рё РїРѕРєР° СЂР°РЅРѕ, РїРѕС‚РѕРјСѓ С‡С‚Рѕ РґР°РЅРЅРѕРµ Р·РЅР°С‡РµРЅРёРµ РїРѕР»СѓС‡РµРЅРѕ РЅРµ РЅР° РІСЃРµС… РґР°РЅРЅС‹С…. РЎРѕР±РµСЂС‘Рј С‚РµРїРµСЂСЊ РґР°РЅРЅС‹Рµ РїРѕ РєСѓСЃРѕС‡РєР°Рј Рё РїСЂРѕРІРµСЂРёРј Р·РЅР°С‡РµРЅРёРµ `r2_score`.

```python
y_test_all = np.hstack([y_test[labels_test == 0], y_test[labels_test == 1]])
y_test_pred_all = np.hstack([y_test_pred_0, y_test_pred_1])

r2_score(y_test_all, y_test_pred_all)
```

РС‚Р°Рє, С‚РµРїРµСЂСЊ РјРѕР¶РЅРѕ СѓС‚РІРµСЂР¶РґР°С‚СЊ, С‡С‚Рѕ РјС‹ РїРѕР»СѓС‡РёР»Рё Р·РЅР°С‡РёС‚РµР»СЊРЅРѕ Р±РѕР»РµРµ РІС‹СЃРѕРєРёР№ РїРѕРєР°Р·Р°С‚РµР»СЊ `r2_score`, РїСЂРёРјРµРЅРёРІ РёРЅС„РѕСЂРјР°С†РёСЋ, РїРѕР»СѓС‡РµРЅРЅСѓСЋ СЃ РїРѕРјРѕС‰СЊСЋ t-SNE Рё K-means.
